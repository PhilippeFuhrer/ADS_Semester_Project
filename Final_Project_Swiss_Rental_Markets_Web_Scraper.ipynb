```python
#!pip install bs4
#!pip install html5lib
#!pip install lxml


from bs4 import BeautifulSoup
import re
import csv
import requests
import warnings
warnings.filterwarnings("ignore",category=UserWarning,module='bs4')

f = open("CSV_Data_File.csv", "w", newline="")
thewriter = csv.writer(f)
thewriter.writerow(["price", "rooms", "squareM", "address", "typus"])

orte = ["zuerich", "luzern", "winterthur", "gallen", "aarau", "zug", "bern", "basel", "solothurn", "lausanne", "genf"]


for ort in orte:

    baseURL="https://www.homegate.ch/mieten/wohnung/region-"+ ort +"/trefferliste?ep="
    page = 1

    while page <=48 :

        source=str(baseURL+str(page))
        soup=BeautifulSoup(source,"lxml")
        r=requests.get(source)
        soup=BeautifulSoup(r.content,"lxml")


        idx=0
        while idx <= 20:
            if (page == 4 and idx > 1) or (page>4):
                x = "ListItem_data"
            else:
                x = "ListItemTopPremium_data"

            boxes=soup.find_all("div",class_=re.compile(x))
            #try:        
             #   price=boxes[idx].find("span", class_=re.compile("ListItemPrice_currency")).text #.findNext("span").text

            try:
                 price=boxes[idx].find("span", class_=re.compile("ListItemPrice_currency")).findNext("span").text
            except:
                price = "null"
            try:
                squareM=boxes[idx].find("span", class_=re.compile("ListItemLivingSpace_value")).text
            except:
                squareM = "null"
            try:
                zimmer=boxes[idx].find("span", class_=re.compile("ListItemRoomNumber_value")).text
            except:
                zimmer = "null"
            try:
                address=boxes[idx].find("span", class_=re.compile("ListItemRoomNumber_value")).findNext("p").text
            except:
                address = "null"
            try:
                typus=boxes[idx].find("span", class_=re.compile("HgBadge_badge")).text
            except:
                typus = "null"

            print(f"{price}||{zimmer}||{squareM}||{address}||{typus}||{idx}||{page}")
            thewriter.writerow([price, zimmer, squareM, address, typus])
            price = "FAIL"
            zimmer = "FAIL"
            address = "FAIL"
            typus = "FAIL"


            idx = idx + 1

        page = page + 1
```
